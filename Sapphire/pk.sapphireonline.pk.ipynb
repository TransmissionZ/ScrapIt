# Scraping Sapphire E-Commerce Website

### Items Scraped:
#####    1) Images of each product
#####    2) Prices
#####    3) Names
#####    4) Descriptions
#####    5) Colors
#####    6) Types

```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import urllib.request
import pandas as pd
```

```python
all_links = []
all_rows = []
```

```python
#Getting HTML of one section:
links = ['https://pk.sapphireonline.pk/collections/unstitched','https://pk.sapphireonline.pk/collections/unstitched?page=2']
for link in links
    uClient = uReq(link)
    base_link = 'https://pk.sapphireonline.pk'
    all_web = uClient.read()
    uClient.close()
    page_soup = soup(all_web, "html.parser")
    product_container = []
    for product in page_soup.findAll("div", {"class":"grid-item col-6 col-md-4 col-xl-3 four-columns"}):
        product_container.append(product)
    #Retrieving All Product Links
    for link in range(len(product_container)):
        x = base_link + product_container[link].a['href']
        all_links.append(x)
```

```python
%%time
x = 0
for link in all_links:
    #print('Working....')
    try:
        uClient = uReq(link)
        all_web = uClient.read()
        uClient.close()
        product_soup = soup(all_web, "html.parser")
        name = product_soup.find("h2", {"itemprop":"name"}).text.strip()
        content = product_soup.find("section", {"id":'content1'})
        detail = content.span.text.strip()
        description = content.p.text.strip()
        content = content.find_all('p')
        #Images
        images = product_soup.find_all("a", {"class":"fancybox"})
        y = 0
        for a in images:
            file_name = str(x) + "." + str(y) + ".jpg"
            #print(a['href'] + file_name)
            y += 1
            urllib.request.urlretrieve("https:" + a['href'], file_name)
            
        x += 1
        print("Images of Product " + str(x) + " Retrieved")
        
        n = -1
        trouser = None
        trouser_Color = None
        trouser_Fabric = None
        shirt_color = None
        shirt = None
        shirt_fabric = None
        dupatta_color = None
        dupatta = None
        dupatta_fabric = None
        if len(content) != 1:
            for p in content:
                n = n+1
                z=0
                if ('Trouser' in p.text):
                    trouser = p.text.replace("Trouser", "",1).strip()
                    trouser_specs = content[n+1].text
                    z = trouser_specs.find("Fabric:")
                    trouser_Color = trouser_specs[7:z].strip()
                    trouser_Fabric = trouser_specs[z:].replace("Fabric:","",1).strip()
                if ('Shirt' in p.text):
                    shirt = p.text.replace("Product Detail","").replace("Unstitched 3-Piece", "").strip()
                    shirt_specs = content[n+1].text
                    z = shirt_specs.find("Fabric:")
                    shirt_color = shirt_specs[7:z].strip()
                    shirt_fabric = shirt_specs[z:].replace("Fabric:", "",1).strip()
                elif ('Dupatta' in p.text):
                    dupatta = p.text.strip()
                    dupatta_specs = content[n+1].text
                    z = dupatta_specs.find("Fabric:")
                    dupatta_color = dupatta_specs[7:z].strip()
                    dupatta_fabric = dupatta_specs[z:].replace("Fabric:", "").strip()
        else:
            content = content[0].text
            Des_slice = content.find("Product Detail")
            description = content[:Des_slice].strip()
            Det_slice = content.find("Shirt")
            detail = content[Des_slice:Det_slice].replace("Product Detail", "").strip()
            Shirt_slice = content.find("Trouser")
            shirt = content[Det_slice:Shirt_slice].replace("Shirt", "",1).strip()
            Trouser_slice = content.find("Dupatta")
            trouser = content[Shirt_slice:Trouser_slice].replace("Trouser", "", 1).strip()
            Dupatta_slice = content.find("Color")
            dupatta = content[Trouser_slice:Dupatta_slice].replace("Dupatta", "", 1).strip()
            Color_slice = content.find("Fabric")
            shirt_color = content[Dupatta_slice:Color_slice].replace("Color:", "").strip()
            Fabric_slice = content.find("Style Note")
            shirt_fabric = content[Color_slice:Fabric_slice].replace("Fabric:", "", 1).strip()
                
                
        prices = product_soup.find("div",{"class":"prices"})('span')
        oldprice = prices[1].text.strip()
        newprice = prices[2].text.strip()

        row = {'Name':name,'Detail':detail,'Description':description,'Old-Price': oldprice, 'New-Price': newprice, 'Shirt':shirt,'Shirt-Color':shirt_color,'Shirt-Fabric':shirt_fabric,'Trouser':trouser,'Trouser-Fabric':trouser_Fabric,'Trouser-Color':trouser_Color,'Dupatta':dupatta,'Dupatta-Color':dupatta_color,'Dupatta-Fabric':dupatta_fabric}
        all_rows.append(row)
        
    except:
        print(link)
        pass
```

```python
products = pd.DataFrame(all_rows, columns = ['Name','Detail','Description','Old-Price','New-Price','Shirt','Shirt-Color','Shirt-Fabric','Trouser','Trouser-Color','Trouser-Fabric','Dupatta','Dupatta-Color','Dupatta-Fabric'])
```
